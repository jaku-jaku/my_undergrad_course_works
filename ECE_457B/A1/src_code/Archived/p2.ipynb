```python
# lib
import numpy as np
import matplotlib.pyplot as plt
import copy
from typing import List, Dict, Optional
```

```python
train_x = [[-1, 1], [1, -1], [1, 1], [-1,-1]]
train_y = [-1, -1, 1, 1]

# Hard-coded weights [biasi, wi1, wi2]
w1 = [-1, 1, 1] 
w2 = [1, 1, 1]
w3 = [0, -1, 1]
```

```python
f_hyper_plane = lambda x, w: (w[0] - w[1] * x)/w[2]

# plot
fig = plt.figure()
ax = plt.axes()
ax.grid(True)
MIN, MAX = -2, 2
ax.set_xlim(MIN, MAX) 
ax.set_ylim(MIN, MAX) 
CLR_LUT = {-1:'r', 1:'b'}
MRK_LUT = {-1:'o', 1:'*'}
for i, y in enumerate(train_y):
    ax.scatter(train_x[i][0], train_x[i][1], c=CLR_LUT[y], marker=MRK_LUT[y])

X = np.linspace(MIN, MAX, 10)
Y1 = f_hyper_plane(X, w1)
Y2 = f_hyper_plane(X, w2)

plt.plot(X, Y1)
plt.plot(X, Y2)

ax.fill_between(X, MIN, Y1, facecolor='blue', alpha=0.1,label="class = 1")
ax.fill_between(X, Y2, MAX, facecolor='blue', alpha=0.1)
ax.fill_between(X, Y1,  Y2, facecolor='red', alpha=0.1, label="class = -1")
ax.legend()
ax.set_aspect(1)
ax.xticks([-1, 1], [-1, 1])
ax.xticks([-1, 1], [-1, 1])
fig.savefig("fig/madaline.png", bbox_inches = 'tight')
```
